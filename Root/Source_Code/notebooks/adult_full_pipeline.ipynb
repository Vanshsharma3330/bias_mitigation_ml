```python
import sys
sys.path.append('..')
# IMPORTS
from src.data_loader import load_adult
from src.preprocessing import encode
from src.models import get_models
from src.fairness_metrics import demographic_parity, equal_opportunity
from src.evaluation import evaluate
from src.visualization import plot_tradeoff, plot_bar

from sklearn.model_selection import train_test_split

# LOAD DATA
df = load_adult()
df = encode(df)

# SPLIT
X = df.drop("income", axis=1)
y = df["income"]
p = df["sex"]

X_train, X_test, y_train, y_test, p_train, p_test = train_test_split(
    X, y, p, test_size=0.2, random_state=42
)

models = get_models()

results = []
labels = []

# BASELINE
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc, f1 = evaluate(y_test, y_pred)
    dpd = demographic_parity(y_pred, p_test.values)
    eod = equal_opportunity(y_test.values, y_pred, p_test.values)

    results.append({"accuracy": acc, "dpd": dpd})
    labels.append(name)

    print(name, acc, f1, dpd, eod)

# PLOTS
plot_tradeoff(results)
plot_bar(results, labels, "dpd", "dpd_comparison.png")
plot_bar(results, labels, "accuracy", "accuracy_comparison.png")
```
